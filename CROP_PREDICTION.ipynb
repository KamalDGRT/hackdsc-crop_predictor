```python
#one hot encoding for categorical variables
sowing_time=pd.get_dummies(crop['Sowing Time'],drop_first=True)
region=pd.get_dummies(crop['Region'],drop_first=True)
soil_type=pd.get_dummies(crop['Soil Type'],drop_first=True)
```

```python
#copying the dataset to new dataset
crop_new=crop.copy()
```

```python
crop_new
```

```python
crop_new.drop(['Sowing Time','Region','Soil Type'],axis=1,inplace=True)
crop_new
```

```python
crop_new=pd.concat([crop_new,sowing_time,region,soil_type],axis=1)
```

```python
crop_new
```

```python
crop_new.to_csv('Agriculture_dsc')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split

X=crop_new.drop("Crop",axis=1)
y=crop_new['Crop']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
rc=RandomForestClassifier(n_estimators=100)
rc.fit(X_train,y_train)
crop_pred=rc.predict(X_test)
```

```python
crop_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score: ",accuracy_score(y_test,crop_pred))
```
